### 本金攤還算式

```python
amount=1000000
years=5
rate=2

amount,years,rate
```

```python
# 固定本金
period=years*12
month_pay=int(amount/period)
month_pay
```

```python
# 月利率
month_rate=rate/100/12
month_rate
```

```python
interest=round(amount*month_rate)
interest
```

### 每個月還款利息

```python
amount=1000000
for i in range(period):
    interest=round(amount*month_rate)
    amount-=month_pay
    if i==(period-1):
        month_pay=month_pay+(amount-month_pay)

    print(f'期別:{i+1} 本金:{month_pay} 利息:{interest} 本利和:{month_pay+interest} 餘額:{amount}')
```
