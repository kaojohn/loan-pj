### 本金攤還算式

```python
amount=1000000
years=5
rate=2

amount,years,rate
```

```python
# 固定本金
period=years*12
month_pay=int(amount/period)
month_pay
```

```python
# 月利率
month_rate=rate/100/12
month_rate
```

```python
interest=round(amount*month_rate)
interest
```

```python
amount=1000000
kane=amount
years=5
rate=2
period=years*12
month_pay=int(amount/period)
month_rate=rate/100/12
alltoto=0

datas=[]
for i in range(period):
    interest=round(amount*month_rate)
    toto=month_pay+interest
    amount-=month_pay

    if i==(period-1):
        datas.append([i+1,month_pay+amount,interest,toto+amount,amount-amount])
        alltoto+=interest
        #print(f'期別:{i+1} 本金:{month_pay+amount} 利息:{interest} 本利和:{toto+amount} 餘額:{amount-amount}')
    else:
        datas.append([i+1,month_pay,interest,toto,amount])
        alltoto+=interest
        #print(f'期別:{i+1} 本金:{month_pay} 利息:{interest} 本利和:{toto} 餘額:{amount}')
    
print(datas)
print(alltoto+kane)
```
